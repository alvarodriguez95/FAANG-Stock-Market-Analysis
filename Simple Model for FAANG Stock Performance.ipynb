```python
import pandas as pd
```

```python
df1 = pd.read_csv('Data/META.csv')
df2 = pd.read_csv('Data/AAPL.csv')
df3 = pd.read_csv('Data/AMZN.csv')
df4 = pd.read_csv('Data/NFLX.csv')
df5 = pd.read_csv('Data/GOOG.csv')
```

```python
# Add a 'Source' column to each dataframe indicating the source
df1['Source'] = 'META'
df2['Source'] = 'AAPL'
df3['Source'] = 'AMZN'
df4['Source'] = 'NFLX'
df5['Source'] = 'GOOG'

# Merge the dataframes using pd.concat()
merged_df = pd.concat([df1, df2, df3, df4, df5])

# Reset the index of the merged dataframe
merged_df = merged_df.reset_index(drop=True)
```
